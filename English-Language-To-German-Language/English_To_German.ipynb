```python
import streamlit as st
from transformers import MarianMTModel, MarianTokenizer

# Load the pre-trained model and tokenizer for English to German translation
model_name = 'Helsinki-NLP/opus-mt-en-de'
model = MarianMTModel.from_pretrained(model_name)
tokenizer = MarianTokenizer.from_pretrained(model_name)

# Define translation function
def translate_text(text):
    # Tokenize the input text
    tokens = tokenizer.encode(text, return_tensors="pt", padding=True)

    # Perform translation
    translation = model.generate(tokens, max_length=100)

    # Decode the translated tokens
    translated_text = tokenizer.decode(translation[0], skip_special_tokens=True)
    return translated_text

# Streamlit app
st.title("English to German Translator")
st.write("Enter English text and get the German translation")

# Text input box for English text
input_text = st.text_area("Enter text in English")

if st.button("Translate"):
    if input_text:
        translated_text = translate_text(input_text)
        st.write(f"Translated text: {translated_text}")
    else:
        st.write("Please enter some text to translate.")
```
